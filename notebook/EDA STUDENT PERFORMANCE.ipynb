```python
numerical_column = [cols for cols in df.columns if df[cols].dtype!='O']
categorical_column= [cols for cols in df.columns if df[cols].dtype=='O']

print('We have {} numerical columns: {}'.format(len(numerical_column),numerical_column))
print('We have {} categorical columns: {}'.format(len(categorical_column),categorical_column))
```

```python
df['Total_score']=df["math_score"]+df["reading_score"]+df["writing_score"]
df["Average"]=df["Total_score"]/3
```

```python
df
```

```python
reading_full_marks=df[df["reading_score"]==100]["Average"].count()
writing_full_marks=df[df["writing_score"]==100]["Average"].count()
math_full_marks=df[df["math_score"]==100]["Average"].count()

print(f"Number of students who got full marks in maths: {math_full_marks}")
print(f"Number of students who got full marks in reading: {reading_full_marks}")
print(f"Number of students who got full marks in writing: {writing_full_marks}")
```

```python
reading_pass_marks=df[df["reading_score"]<=40]["Average"].count()
writing_pass_marks=df[df["writing_score"]<=40]["Average"].count()
math_pass_marks=df[df["math_score"]<=40]["Average"].count()

print(f"Number of students who got passing marks or failed in maths: {math_pass_marks}")
print(f"Number of students who got passing marks or failed in reading: {reading_pass_marks}")
print(f"Number of students who got passing marks or failed in writing: {writing_pass_marks}")
```

```python
#multivariate analysis using pieplot
plt.subplots(1,5,figsize=(30,12))

plt.subplot(1,5,1)
size=df["gender"].value_counts()
label="Female","Male"
color=['red','blue']

plt.pie(x=size,labels=label,colors=color,autopct=".%2f%%")
plt.title("Gender")

plt.subplot(1,5,2)
size=df["race_ethnicity"].value_counts()
label="Group A","Group B","Group C","Group D","Group E"
color=['red','blue','green','yellow','pink']

plt.pie(x=size,labels=label,colors=color,autopct=".%2f%%")
plt.title("Race/Ethnicity")

plt.subplot(1,5,3)
size=df["parental_level_of_education"].value_counts()
label='Some College', "Associate's Degree",'High School','Some High School',"Bachelor's Degree","Master's Degree"
color=['red','blue','purple','orange','pink','green']

plt.pie(x=size,labels=label,colors=color,autopct=".%2f%%")
plt.title("Parent's Education")

plt.subplot(1,5,4)
size=df["lunch"].value_counts()
label='standard', 'free/reduced'
color=['red','blue']

plt.pie(x=size,labels=label,colors=color,autopct=".%2f%%")
plt.title("Lunch")

plt.subplot(1,5,5)
size=df["test_preparation_course"].value_counts()
label='none', 'completed'
color=['red','blue']

plt.pie(x=size,labels=label,colors=color,autopct=".%2f%%")
plt.title("Test Preparation")

plt.tight_layout()
plt.grid()

plt.show()
```
